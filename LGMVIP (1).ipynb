**Importing required libraries**

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
def show(title,img,cmap='RdBu'):
    plt.imshow(img,cmap=cmap)
    plt.title(title)
```

**Reading the image**

```python
image=plt.imread("/content/flower.png")
show("flower",image,'RdBu')
```

**Converting image into grayscale image**

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
show("gray_image",gray_image,'gray')
```

**Invert the grayscale image**

```python
inverted_image =255-gray_image
show("inverted_image",inverted_image,'gray')
```

**Blur the inverted image**

```python
blurred_image =cv2.GaussianBlur(inverted_image,(21,21),0)
show("blurred_image",blurred_image,'gray')
```

**Inverting the blurred image**

```python
inverted_blurred_image = 255 - blurred_image
show("inverted_blurred_image",inverted_blurred_image,'gray')
```

**Create pencil sketch image**

```python
pencil_sketch=cv2.divide(gray_image,inverted_blurred_image,scale=256.0)
show("pencil_sketch",pencil_sketch,'gray')
```
